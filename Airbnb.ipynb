**$~~$9. Price Distribution**

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, bins=50,kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

Key Observation:

The distribution of prices is right-skewed, with most listings priced between $40000 and $50000.

A small number of listings have extremely high prices (outliers), suggesting luxury properties or mansions.


**10.$~~$Room Type vs Price**

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='room_type', y='price', data=df,hue='room_type')
plt.title('Room Type vs Price')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```

Key Observation:

Entire Home/Apt listings are generally more expensive than Private Rooms or Entire Rooms under $6000-10000,

The median price for Entire Home/Apt is around $2000  while Private Rooms tend to be under $8000.



**11.$~~$Price vs Number of Reviews**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='number_of_reviews', y='price', data=df, hue='room_type')
plt.title('Price vs Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.show()
```

Key Observation:

Higher-priced listings tend to have fewer reviews. This suggests that they may be more niche or less frequently booked.
Lower-priced listings have a higher volume of reviews, indicating more frequent bookings.

**12.$~~$Availability vs Price**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='availability_365', y='price', data=df,hue='room_type')
plt.title('Availability vs Price')
plt.xlabel('Availability (days/year)')
plt.ylabel('Price')
plt.show()
```

Key Observation:

Listings that are more available (closer to 365 days) tend to have lower prices on average, as they might be budget-friendly or in less popular areas.
Expensive listings are often less available, suggesting that they may be premium properties or located in high-demand areas.

**13.$~~$Neighbourhood Group vs Price**

```python
plt.figure(figsize=(10,6))
sns.barplot(x='neighbourhood_group', y='price', data=df, hue='room_type')
plt.title('Neighbourhood Group vs Price')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
plt.show()
```

Key Observation:

Manhattan and Brooklyn have the highest median prices, reflecting their popularity and demand.
Some neighbourhood groups show a wide range of prices, indicating variability in property types and luxury levels.

**14.$~~$Number of Reviews vs Reviews Per Month**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='number_of_reviews', y='reviews_per_month', data=df)
plt.title('Number of Reviews vs Reviews Per Month')
plt.xlabel('Number of Reviews')
plt.ylabel('Reviews Per Month')
plt.show()
```

Key Observation:

Listings with more reviews tend to have a more consistent review rate per month.
Properties with fewer reviews may experience sporadic review submissions, potentially indicating infrequent bookings or seasonal properties.


**15.$~~$Room Type Distribution**

```python
plt.figure(figsize=(10,6))
sns.countplot(x='room_type', data=df,hue='room_type')
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

Key Observation:

Entire Home/Apt is the most frequent room type, making up around 60% of the listings.
Private Rooms are the next most common, with fewer Shared Rooms being available

**16.$~~$Host Listings Count Distribution**

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, bins=30, kde=True)
plt.title('Host Listings Count Distribution')
plt.xlabel('Number of Listings')
plt.ylabel('Frequency')
plt.show()
```

Key Observation:

Most hosts list only 1 or 2 properties.
A small number of hosts (superhosts) have a high number of listings, which indicates professional property managers.

**17.$~~$Pairpot for price,minimum_nights,number_of_reviews,availability_365**



```python
sns.pairplot(data=df,vars=['price', 'minimum_nights', 'number_of_reviews',  'availability_365'],hue='room_type')
```

Key Observations:

Room type (e.g., Entire Home/Apartment) is strongly associated with higher prices and possibly more reviews.

Listings with lower minimum nights tend to be more available throughout the year (e.g., Private Rooms or Shared Rooms).

There are clear price and review patterns that differ significantly across room types, with Entire Home/Apartment generally more expensive and having more reviews.




**18.$~~$ Correlation Heatmap**

```python
correlation_matrix = df[['price', 'minimum_nights', 'number_of_reviews','reviews_per_month', 'availability_365']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of price,mminimum_nights,number_of_reviews,reviews_per_month,availability_365')
plt.show()
```

Key Observation:
"Entire home/apt" listings tend to cluster in more affluent, residential, or tourist-centric neighborhoods, reflecting the demand for privacy and comfort. In contrast, "Shared room" and "Private room" listings are more evenly distributed, likely catering to budget-conscious travelers who prioritize affordability over space or privacy. This suggests that higher-end accommodations are concentrated in specific, desirable areas, while more affordable options are available throughout the city.



**19.$~~$Pie Chart: Distribution of Room Types**

```python
room_type_counts = df['room_type'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', startangle=90, colors=['green', 'purple', 'yellow'])
plt.title('Distribution of Room Types')
plt.show()
```


This pie chart shows the percentage distribution of room types. It helps identify the most common types of accommodations .

**20.$~~$Neighborhood Group Distribution**
